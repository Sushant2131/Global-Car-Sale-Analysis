```python
df.columns
# drop some columns
columns_drop=['Nickname', 'Car Gender','Buzzword','City']
df.drop(columns_drop,axis=1,inplace=True)
```

```python
# rename some columns
df.rename(columns={'Make':'Company','Model':'Car_Model','Discount':'Discount %','5-yr Depreciation':'5-yr Depreciation %','Top Speed':'Top Speed kmph'},inplace=True)
```

```python
# convert the price into float to int
column_convert=['Sale Price','Resell Price','Top Speed kmph']
df[column_convert] = df[column_convert].round().astype('int64')
```

```python
# convert the discount into normal
df['Discount %']=df['Discount %']*100
```

```python
# convert the Depreciation into normal %
df['5-yr Depreciation %']=df['5-yr Depreciation %']*100
```

```python
import pycountry
# Function to get ISO country code
def get_iso_country_code(Country):
    try:
        country = pycountry.countries.lookup(Country)
        return country.alpha_2
    except LookupError:
        return None

# Add ISO country code column based on country name
df['Country Code'] = df['Country'].apply(get_iso_country_code)

```
